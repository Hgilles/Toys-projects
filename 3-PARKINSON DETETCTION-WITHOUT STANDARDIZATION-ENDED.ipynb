# 3-PARKINSON DETECTION

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv(r'C:\parkinsons.data')
```

```python
df.head()
```

```python
df.info()
```

```python
df0=df.drop('name', axis=1)
df0.isnull().sum()
lab=df0['status']
df9=df0.drop('status', axis=1)
lab.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df9.hist(figsize=(18,22), bins=40, xlabelsize=8, ylabelsize=8)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
xtrain, xtest, ytrain,  ytest=train_test_split(df9,lab, test_size=0.2)
```

```python
xtrain
```

```python
gnb=GaussianNB()
model=gnb.fit(xtrain,ytrain)
```

```python
g=gnb.predict(xtest)
g
```

```python
from sklearn.metrics import accuracy_score 
```

```python
accuracy_score(ytest, g)
```

```python
dec=DecisionTreeClassifier()
mod0=dec.fit(xtrain,ytrain)
```

```python
a=mod0.predict(xtest)
```

```python
accuracy_score(ytest, a)
```

At this moment DecisionTreeClassifier look on the right way to be the ideal model for parkinson detection.
